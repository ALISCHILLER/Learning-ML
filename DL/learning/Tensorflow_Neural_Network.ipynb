```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

# data definition

```python
# y = 3 * x - 5
# داده‌ها را به آرایه‌های NumPy تبدیل کنید
x = np.array([-1,-2,0,1,2,5,7] , dtype=float)
y = x * 3 - 5
```

```python
x
```

```python
y
```

# MODEL DEFINITION

```python
model = keras.Sequential([
    keras.layers.Dense(units=1, activation=None, input_shape=[1])
])
```

```python
# (y - y_pred)^2
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python

model.summary()
```

```python
# داده‌ها را به عنوان ورودی و خروجی به تابع fit بدهید
model.fit(x, y, batch_size=1, epochs=500)
```

## test

```python
# y = 3 * x - 5
x_test = tf.constant([-4,11,20], dtype=float)
y_test = x_test * 3 - 5
```

```python
y_test.numpy()
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```
