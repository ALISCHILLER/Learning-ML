```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# data 
# y = 2*x -1

x = np.array([1,2,3,4]).reshape(-1,1)
y = np.array([1,3,5,7])
x.shape
```

```python
x
```

```python
plt.plot(x,y,'*')
plt.show()
```

```python
# model 
import LinearRegression
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
# y = m * x +  c
m= model.coef_[0]
c = model.intercept_
print(m , c)
```

```python
plt.plot(x,y,'*')
plt.plot(x , model.predict(x))
plt.show()
```

```python
result = model.predict([[5]])
result
```

```python
years_of_expreience = np.array([0,1,2,3,4,6,8,10]).reshape(-1, 1)
salaries = np.array([30000,40000,55000,60000,70000,80000,85000,87000])
```

```python
plt.plot(years_of_expreience, salaries , '*')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=3)
model.fit(years_of_expreience,salaries)
y_predict = model.predict(years_of_expreience)
plt.plot(years_of_expreience, salaries, '*')
plt.plot(years_of_expreience, y_predict)
plt.show()
```
