```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [4,7,8,9,3,2,1,5,5]
plt.plot(x,y,marker='o',color="green")
plt.title("test")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = ["home","car","work","gem"]
y = [12,33,3,6]
plt.bar(x,y , color="red")
plt.title("test2")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = ["home","car","work","gem"]
y = [12,33,3,6]

plt.pie(y , labels=x,autopct="%1.1f%%",colors=["red","green","blue","orange"])
plt.title("test3")
plt.show()
```
